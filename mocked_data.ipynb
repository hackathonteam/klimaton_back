```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
deklarowane = pd.read_excel("data/06deklarowane_with_history.xlsx")
zuzycie = pd.read_excel("data/01zuzycie.xlsx")
```

```python
deklarowane = deklarowane.dropna(axis=0)
zuzycie = zuzycie.dropna(axis=0)
```

```python
# divide values to account for zeroes
for i in range(145):
    if zuzycie.iloc[i, 6] != 0:
        for j in range(17):
            temp = zuzycie.iloc[i, 6 + j * 2]
            zuzycie.iloc[i, 6 + j * 2] = temp / 2
            zuzycie.iloc[i, 6 + j * 2 + 1] = temp / 2
    else:
        for j in range(17):
            temp = zuzycie.iloc[i, 6 + j * 2 + 1]
            zuzycie.iloc[i, 6 + j * 2] = temp / 2
            zuzycie.iloc[i, 6 + j * 2 + 1] = temp / 2
    zuzycie.iloc[i, 40] = zuzycie.iloc[i, 39]
```

```python
dekl_hist_0 = deklarowane.iloc[0,9:]
zuz_hist_0 = zuzycie.iloc[0,6:]
print(dekl_hist_0.shape)
print(zuz_hist_0.shape)
print(type(dekl_hist_0.iloc[::-1].values))
```

```python
dates = [str(x.month) + "-" + str(x.year) for x in dekl_hist_0.index]
dates.reverse()
```

```python
indices = np.arange(len(dates[:6]))

width = 0.20

plt.bar(indices, dekl_hist_0.iloc[::-1][:6], width=width)

# Offsetting by width to shift the bars to the right
plt.bar(indices + width, zuz_hist_0.iloc[::-1][:6], width=width)


# Displaying year on top of indices
plt.xticks(ticks=indices, labels=dates[:6], rotation=90)


plt.xlabel("data")
plt.ylabel("deklarowane ścieki")
```
